```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def get_vif_factor( X ):
    X_matrix = X.to_numpy()
    vif = [ variance_inflation_factor(X_matrix, i) for i in range(X_matrix.shape[1])]
    vif_factors = pd.DataFrame()
    vif_factors['column'] = X.columns
    vif_factors['VIF'] =vif
    return vif_factors
```

```python
vif_factor = get_vif_factor(X[X.columns[1:]])
vif_factor
```

```python
vif_columns_large = vif_factor[vif_factor.VIF > 4].column
```

```python
import scipy as sp
fig, ax = plt.subplots(figsize=(6, 8), dpi = 100)
_ = sp.stats.probplot(sign_model.resid, plot=ax)
```

```python
def draw_pp_plot( model, title ):
    probplot = sm.ProbPlot( model.resid )
    plt.figure( figsize=(8, 6) )
    probplot.ppplot( line="45" )
    plt.title(title)
    plt.show()
```

```python
draw_pp_plot( sign_model, 'Prob-Plot' ) 
```

```python
def get_standardized_values(vals):
    return (vals-vals.mean())/vals.std()
```

```python
def plot_resid_fitted(fitted, resid, title):
    plt.scatter(get_standardized_values(fitted), get_standardized_values(resid))
    plt.title(title)
    plt.xlabel("Fitted Values")
    plt.ylabel("Pred Values")
    plt.axhline(y = 0, color = 'r', ls = '--')
    plt.show()
```

```python
plot_resid_fitted(sign_model.fittedvalues, sign_model.resid, 'Residual Plot')
```
