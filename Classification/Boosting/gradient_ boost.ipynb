```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Data Science\udemy\Notebooks\DATA\mushrooms.csv")
```

```python
df.head()
```

```python
X = df.drop('class', axis = 1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, test_size = 0.1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 128],
    'learning_rate': [0.1, 0.05, 0.2],
    'max_depth': [3, 4, 5]
}
```

```python
gb_model = GradientBoostingClassifier()
```

```python
grid = GridSearchCV(gb_model, param_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
plot_confusion_matrix(grid, X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
grid.best_estimator_
```

```python
feat_impor = grid.best_estimator_.feature_importances_
```

```python
imp_feat = pd.DataFrame(index = X.columns, data= feat_impor, columns = ['Importance'])
```

```python
imp_feat = imp_feat[imp_feat['Importance'] > 0.0005]
```
