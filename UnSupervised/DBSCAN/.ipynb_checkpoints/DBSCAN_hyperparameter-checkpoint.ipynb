```python
sns.scatterplot(data = two_blobs, x = 'X1',y='X2')
plt.show()
```

```python
sns.scatterplot(data = two_blobs_outliers, x = 'X1',y='X2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN()
```

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data = data, x='X1', y='X2', hue=labels, palette = 'Set1')
```

```python
display_categories(dbscan, two_blobs)
```

```python
display_categories(dbscan, two_blobs_outliers)
```

```python
dbscan = DBSCAN(eps=0.001)
display_categories(dbscan, two_blobs_outliers)
```

```python
dbscan = DBSCAN(eps=10)
display_categories(dbscan, two_blobs_outliers)
```

```python
dbscan = DBSCAN(eps=1)
display_categories(dbscan, two_blobs_outliers)
```

```python
np.sum(dbscan.labels_ == -1)
```

```python
np.sum(dbscan.labels_ == -1) * 100 / len(dbscan.labels_)
```

```python
outlier_per = list()
```

```python
num_of_out = list()
```

```python
for i in np.arange(1, 100):
    dbscan = DBSCAN(eps=i)
    dbscan.fit(two_blobs_outliers)
    num_of_out.append(np.sum(dbscan.labels_ == -1))
    outlier_per.append(np.sum(dbscan.labels_ == -1) * 100 / len(dbscan.labels_))
```

```python
sns.lineplot(x=np.linspace(0.001, 10, 200), y = num_of_out)
plt.xlim(0, 2)
plt.ylim(0, 10)
plt.hlines(y=3, xmin = 0, xmax = 2, color = 'red')
plt.show()
```

```python
sns.lineplot(x=np.linspace(0.001, 10, 200), y = outlier_per)
plt.xlim(0, 2)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.75)
display_categories(dbscan, two_blobs_outliers)
```

```python
sns.lineplot(x = np.arange(1, 100), y = outlier_per)
plt.show()
```

```python
outlier_per = list()
num_of_out = list()
for i in np.arange(1, 100):
    dbscan = DBSCAN(min_samples=i)
    dbscan.fit(two_blobs_outliers)
    num_of_out.append(np.sum(dbscan.labels_ == -1))
    outlier_per.append(np.sum(dbscan.labels_ == -1) * 100 / len(dbscan.labels_))
```

```python
sns.lineplot(x=np.arange(1, 100), y = outlier_per)
```

```python
num_dim = two_blobs_outliers.shape[1]
dbscan = DBSCAN(min_samples = 2*num_dim,eps=0.75)
display_categories(dbscan, two_blobs_outliers)
```
