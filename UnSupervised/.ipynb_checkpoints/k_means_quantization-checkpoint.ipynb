```python
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import matplotlib.image as pltimg
```

```python
img_arr = pltimg.imread(r'D:\Data Science\udemy\Notebooks\DATA\palm_trees.jpg')
```

```python
img_arr
```

```python
plt.figure(dpi=200)
plt.imshow(img_arr)
plt.show()
```

```python
(h, w, c) = img_arr.shape
```

```python
img_as_arr = img_arr.reshape(h*w, c)
```

```python
len(img_as_arr.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(img_as_arr)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
quad_img = np.reshape(rgb_codes[labels], (h,w,c))
```

```python
plt.figure(dpi=200)
plt.imshow(quad_img)
plt.show()
```
