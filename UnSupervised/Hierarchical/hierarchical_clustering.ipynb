```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'D:\Data Science\udemy\Notebooks\DATA\cluster_mpg.csv')
```

```python
df_dum = pd.get_dummies(df.drop('name', axis = 1))
```

```python
df_dum
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
S_data = mms.fit_transform(df_dum)
```

```python
S_data
```

```python
s_df = pd.DataFrame(S_data, columns = df_dum.columns)
```

```python
s_df
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(s_df, cmap = 'viridis', annot = True)
plt.show()
```

```python
len(df)
```

```python
plt.figure(figsize=(15, 8))
sns.clustermap(s_df, cmap = 'viridis', annot = True, row_cluster= False)
plt.show()
```

```python
sns.heatmap(s_df.corr(), annot = True, cmap = 'Blues')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.clustermap(s_df, cmap = 'viridis', annot = True, col_cluster=False)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters = 4)
```

```python
cluster_labels = model.fit_predict(s_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(data= df, x = 'mpg', y = 'weight', hue = cluster_labels)
plt.show()
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(data= df, x = 'mpg', y = 'horsepower', hue = cluster_labels, palette = 'viridis')
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(s_df)
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
from scipy.cluster import hierarchy
```

```python
linkage_m = hierarchy.linkage(model.children_)
```

```python
linkage_m
```

```python
plt.figure(figsize=(20, 10))
den = dendrogram(linkage_m, truncate_mode='lastp', p = 11)
plt.show()
```
